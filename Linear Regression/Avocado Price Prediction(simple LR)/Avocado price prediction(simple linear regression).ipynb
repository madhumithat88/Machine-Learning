```python
import matplotlib.pyplot as plt
import pandas as pd
import pylab as pl
import numpy as np
%matplotlib inline
```

```python
df= pd.read_csv(r'C:\Users\HP\Downloads\avocado.csv')
```

```python
df.describe()
```

```python
df1=df.replace('?',np.NaN)
```

```python
df=df1.dropna(subset=["Total Volume"], axis=0)
df.head(10)
```

```python
df.dtypes
```

```python
df.info()
```

```python
df= df.rename(columns = {'Total Volume': 'TotalVolume'}, inplace = False)
df= df.rename(columns = {'Total Bags': 'TotalBags'}, inplace = False)
```

```python
plt.scatter(df.AveragePrice, df.TotalVolume,  color='blue')
plt.xlabel("Average Price")
plt.ylabel("Total Volume")
plt.show()
```

```python
plt.scatter(df.AveragePrice, df.TotalBags,  color='blue')
plt.xlabel("Average Price")
plt.ylabel("Total Bags Sold")
plt.show()
```

```python
cdf = df[['AveragePrice', 'TotalBags']]
cdf.head(9)
```

```python
cdf.dropna(axis=1, how='all')
cdf.dropna(axis=1, thresh=1)
```

```python
msk = np.random.rand(len(df)) < 0.8
train = cdf[msk]
test = cdf[~msk]
```

```python
plt.scatter(train.AveragePrice, train.TotalBags,  color='blue')
plt.xlabel("Average price of each avacado")
plt.ylabel("Total no. of bags sold")
plt.show()
```

```python
from sklearn import linear_model
regr = linear_model.LinearRegression()
train_x = np.asanyarray(train[['AveragePrice']])
train_y = np.asanyarray(train[['TotalBags']])
regr.fit (train_x, train_y)
print ('Coefficients: ', regr.coef_)
print ('Intercept: ',regr.intercept_)
```

```python
plt.scatter(train.AveragePrice, train.TotalBags,  color='blue')
plt.plot(train_x, regr.coef_[0][0]*train_x + regr.intercept_[0], '-r')
plt.xlabel("Average price of each avocado")
plt.ylabel("Total no. of bags sold")
```

```python
cdf.dtypes
```

```python
test_x=np.asanyarray(test[['AveragePrice']])
test_y=np.asanyarray(test[['TotalBags']])
```

```python
accuracy=regr.score(test_x, test_y)
"Accuracy: {}%".format(int(round(accuracy*100)))
```
